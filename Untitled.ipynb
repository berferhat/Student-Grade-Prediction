```python
X_train, X_test, y_train, y_test = train_test_split(student, labels,
                                                    test_size = 0.25, 
                                                    random_state = 42)
```

```python
# MAE ve RMSE Hesaplama
def evaluate_predictions(predictions, true):
    mae = np.mean(abs(predictions - true))
    rmse = np.sqrt(np.mean((predictions - true) ** 2))
    
    return mae, rmse
```

```python
# Medyan bulma
median_pred = X_train["G3"].median()

# Medyanları içeren bir liste
median_pred_list = [median_pred for _ in range(len(X_test))]

# gerçek G3 değerlerini saklayalım
true_G3 = X_test["G3"]
```

```python
# naive base değerleri
mb_mae, mb_rmse = evaluate_predictions(median_pred_list, true_G3)

print('Median Baseline  MAE: {:.4f}'.format(mb_mae))
print('Median Baseline RMSE: {:.4f}'.format(mb_rmse))
```

```python
# Eğitim ve Test verilerini birkaç ML modeli ile test edelim

def evaluate(X_train, X_test, y_train, y_test):
    # Model isimleri
    model_isim_listesi = ["Linear Regression", "Random Forest", 
                          "Extra Trees", "SVM", "Gradient Boosted"]
    X_train = X_train.drop("G3", axis = "columns")
    X_test = X_test.drop("G3", axis = "columns")

    # Modelleri tanımlama
    model1 = LinearRegression()
    model2 = RandomForestRegressor(n_estimators = 100)
    model3 = ExtraTreesRegressor(n_estimators = 100)
    model4 = SVR(kernel = "rbf", degree = 3, C = 1.0, gamma = "auto")
    model5 = GradientBoostingRegressor(n_estimators = 50)
    
    # Sonuçlar için DataFrame
    results = pd.DataFrame(columns = ["mae", "rmse"], index = model_isim_listesi)
    
    
    # Modellerin eğitilmesi
    for i, model in enumerate([model1, model2, model3, model4, model5]):
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)
        
        
        # Metrics
        mae = np.mean(abs(predictions - y_test))
        rmse = np.sqrt(np.mean((predictions - y_test) ** 2))
        
        
        # Model sonuçlarının DataFrame aktarılması
        model_adi = model_isim_listesi[i]
        results.loc[model_adi, :] = [mae, rmse]
    
    
    return results
```

```python
from sklearn.metrics import accuracy_score 
#X_train, X_test, y_train, y_test
model1 = RandomForestRegressor(n_estimators=100)
model1.fit(X_train, y_train)
predictions = model1.predict(X_test)
mae = np.mean(abs(predictions - y_test))
rmse = np.sqrt(np.mean((predictions - y_test) ** 2))
print(mae, " - ", rmse) 
```

```python
results = evaluate(X_train, X_test, y_train, y_test)
results
```

```python
plt.figure(figsize=(12, 8))

# Root mean squared error
ax =  plt.subplot(1, 2, 1)
results.sort_values('mae', ascending = True).plot.bar(y = 'mae', color = 'b', ax = ax, fontsize=20)
plt.title('Model Mean Absolute Error', fontsize=20) 
plt.ylabel('MAE', fontsize=20)

# Median absolute percentage error
ax = plt.subplot(1, 2, 2)
results.sort_values('rmse', ascending = True).plot.bar(y = 'rmse', color = 'r', ax = ax, fontsize=20)
plt.title('Model Root Mean Squared Error', fontsize=20) 
plt.ylabel('RMSE',fontsize=20)

plt.show()
```
